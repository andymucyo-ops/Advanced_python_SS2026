#### Exercise 2: Filtering *(2 points)*

Implement a function `odd_numbers` that accepts a list of integers and returns a list containing only the odd numbers of this list.

Example: `print(odd_numbers([1, 2, 3]))` should output `[1, 3]`

````python
def odd_numbers(int_list: list[int]) -> list[int]:
    #return a list containing all element of int_list which are odd using list comprehension
    return [i for i in int_list if i%2 != 0]

test_list: list[int] = [1,2,3,4,5]
print(odd_numbers(test_list))
````

#### Exercise 5: Number divisors *(4 points)*

A divisor of a number, is another number that divides the first number without reminder.

Example: 4 is a divisor of 20. 1 and the number itself are always divisors. 

Write a function `find_divisors` that takes a number in input and returns *all* its divisors.

Example:
```python 
find_divisors(20)  # returns [1, 2, 4, 5, 10, 20]
find_divisors(7)  # returns [1, 7]
```

````python
def find_divisors(number: int) -> list[int]:
    """
    we first create a list containing all potential dividers, by creating a list from 1 until the number analyzed 
    (number + 1 in the range function, because it by default does not include the upper value)

    then using list comprehension we return a sublist of the potential_div list where the result of the float division
    is the same as the result of the floor division for number/div 
    """
    potential_div: list[int] = list(range(1,number+1))
    return [div for div in potential_div if (number/div)==(number//div)]

print(find_divisors(20))
print(find_divisors(7))
````

Implement a function `find_common_divisors` that accepts to integers and returns a sorted list of common divisors (high divisors first).

Hints: Use the function `find_divisors` from above. Sets can be helpful to implement list intersections.

````python
def find_common_divisors(a:int, b:int) -> list[int]:
    """"
    using list comprehension and taking advantage of the returns of the fund_divisors function
    we create a list contains all dividers of integer a, that are also dividers of integer b 
    """
    return [num for num in find_divisors(a) if num in find_divisors(b) ]


print(find_common_divisors(20,10))
````
