Explain why both lists contain 4, even though we do not have code directly involving `list_1`.

**Your answer**: In Python variable names "point" towards an objects content. Therefore multiple variable names can point towards the same object in memory. In the example above list_1 and list_2 point to the same object ( which is a list with the values 1,2,3 ). If the object is accessed by one of the variable names pointing towards it and modified, of course the other variable name will display the modified object too.

Demonstrate your explanation by using `id(...)` and comments.

````python
# TODO
list_1 = [1, 2, 3]
list_2 = list_1

print("List 1:", list_1)
print("List 2:", list_2)

#print the id of both lists
print(f'id list_1: {id(list_1)}, id list_2: {id(list_2)}')

list_2.append(4)

print("List 1:", list_1)
print("List 2:", list_2)

#print the ids of both lists, notice that they didn't change for neither of the two.
print(f'id list_1: {id(list_1)}, id list_2: {id(list_2)}')
````

````python
import random

def slot_machine():
    #this creates and returns a list of 3 random numbers, similar to a slot machine or rolling 3 dice
    lucky_numbers = [random.randint(0,9) for x in range(3)]
    return lucky_numbers

#=================================================================================
def gamble(fee, verbose = False):

    #initialize spending
    spending = 0

    #adjust spending
    spending = spending - fee

    #generate 3 lucky numbers and store them in a list
    lucky_numbers = slot_machine()

    #initialize loop
    while lucky_numbers != [6,6,6]:
        
        #print spending and lucky numbers
        if(verbose == True):
            print( f'your numbers: {lucky_numbers}, your spending: {spending}')
        
        #generate again 3 lucky numbers and store them in a list
        lucky_numbers = slot_machine()

        #adjust the spending
        spending = spending - fee

    if(verbose == True):
        print(f'your numbers: {lucky_numbers}. Congratulation you won!')
    return spending

#==================================================================================
def simulate(n,fee) :
    #create a list of simulation for the number a number of samples n 
    avg_spendings = [gamble(fee) for x in range(n)]

    return sum(avg_spendings)/len(avg_spendings)


#simulation

print("The average spending for 500 samples with 5 chf fees is: ", simulate(500,5) )
````
