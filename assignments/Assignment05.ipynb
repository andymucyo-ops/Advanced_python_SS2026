```python
# TODO
# sort accoring to exercise
sorted_mountains_df = mountains_df.sort_values("first_ascent", ascending=False)

# print output
print(sorted_mountains_df)

# showing difference between loc and iloc
print(f" loc: {sorted_mountains_df.loc[0]}")
print(f" iloc: {sorted_mountains_df.iloc[0]}")
```

**Your answer**: loc will return the value with the index given (index 0 is Mont Blanc), that is called label based indexing. iloc on the other hand returns the value thats currently on the given index (index 0 is Piz Sena, top of the sorted dataframe), which is called positional based indexing.

The dataset for this exercise contains the counts of first names of the population by gender for the years 2024 and 2016.

Note: For privacy reasons, names with counts lower than 3 were counted as `0`.

1. Load the CSV file `data/csv/names.csv` as a dataframe `first_names_df`.

```python
# TODO
# load data ("../" had to be added) 
first_names_df = pd.read_csv("../data/csv/names.csv")

# print the df
print(first_names_df)
```

2. Create a new column `total` containing the sum of the male and female counts.

```python
# TODO
# adding a column called total (sum of male and female counts)
first_names_df["total"] = first_names_df["female"] + first_names_df["male"]

# print outcome
print(first_names_df)
```

3. For the year 2024, calculate the the percentage of distinct names whose total count is below 100 (i.e., percentage of rare first names).

```python
# TODO
# create df with all names from 2024
df_2024 = first_names_df[first_names_df["year"] == 2024]

# sum name that have an abundance of less than 100
rare = (df_2024["total"] < 100).sum()

# total lenght of df_2024
total_names = len(df_2024)

# calculating percentage
percentage = rare / total_names * 100

# print percentage
print(f"{percentage:.2f}%")
```

4. Which 5 names had the biggest increase in 2024 compared to 2016?

Hint: create a pandas Series per year. Exercise 2 might help you.

```python
# TODO 

# create a names series for the year 2024 and 2016 
names_2024 = first_names_df[first_names_df["year"] == 2024].set_index("name")["total"]
names_2016 = first_names_df[first_names_df["year"] == 2016].set_index("name")["total"]

# take difference
increase = names_2024 - names_2016

# filter for top 5 most abundant names
top5 = increase.nlargest(5)

# print outcome
print(top5)
```
