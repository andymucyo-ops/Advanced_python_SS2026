An important software development principle is **DRY** (Don't Repeat Yourself).

The final output must exactly match the current output.

#### Exercise 7: Sets *(2 points)*

Create code that generates **3 random integers from 1 to 6** using a loop. Add each generated integer to a set named `occurred_numbers`. Finally, print the set.

Hint: Generate the random integer with `random.randint(1, 6)`.

````python
import random

#create the empty set that will receive the random numbers 
occurred_numbers: set = set()

# iterate 3 times to generate a random number and add it to the occurred_numbers set
for i in range(3):
    occurred_numbers.add(random.randint(1,6))

# print the final set
print(occurred_numbers)   
````

Consider a simple game: if the generated numbers are 1–2–3 (in this order), the player wins. To implement this game, what would you suggest changing in the code so far?

**Your answer:** The occurred_number should be used to verify if a number has already appeared, in order that generated numbers are added to a results list, only if they are absent from occurred_numbers. They need to be added to a list, since the order matters. exemple below

````python
import random

# create the empty set that will receive the random numbers that were already generated
# and draw list that will receive the random numbers, like a lottery draw 
occurred_numbers: set = set()
draw: list = []

# Loop until the length of the draw reaches 3 items, adding an randomly generated int to the draw list 
# only if this number has not occurred yet
while len(draw) < 3:
    new_num: int = random.randint(1,6)
    if new_num not in occurred_numbers:
        occurred_numbers.add(new_num)
        draw.append(new_num)
    continue #goes to next iteration of the loop if the number has already occurred 

#print user's draw
print("Your final draw is:", draw)

# Output to print if win or lose 
if draw == [1,2,3]:
    print("You win!")
else:
    print("You lose!")

# print the final set
# print(occurred_numbers)
# print(draw) 
````

#### Exercise 8: Loops *(2 points)*

Implement a function `average_word_length`, that accepts a list of strings and calculates the average length of all the words given as input.

Example: `print(average_word_length(["moon", "sun", "stars"]))` should output `4.0`

````python
def average_word_length(words: list[str]) -> float:
    # get a list containing the length of each words of the list by applying the len function on every
    # items of the list
    word_lengths: list[int] = list(map(len, words))
    return sum(word_lengths)/len(words) # return the total of the word lengths divided by the length of the list

test_list: list[str]= ["moon", "sun", "stars"]

print("Average length of words:", average_word_length(test_list))
````
